## Exercise 1

**Follow along with these steps**

* Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax.
* Plot (x,y) on that axes and set the labels and titles to match the plot below:

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
plt.show()
```

## Exercise 3

**Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]**

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax = fig.add_axes([0.2,0.5,0.4,0.4])
plt.show()
```

**Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 100, 20)
z = x**2
y = 2*x
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(x, z)
ax.set_xlabel('X')
ax.set_ylabel('Z')

ax_inset = plt.axes([0.2, 0.5, 0.4, 0.3])  # [left, bottom, width, height]
ax_inset.plot(x, y)
ax_inset.set_xlim(20, 22)
ax_inset.set_ylim(30, 50)
ax_inset.set_xlabel('X')
ax_inset.set_ylabel('Y')
ax_inset.set_title('zoom')

plt.show()
```

## Exercise 4

**Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=1, ncols=2)

plt.show()
```

**Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
fig = plt.figure()
plt.subplots(nrows=1, ncols=2)
plt.subplot(1, 2, 1)
plt.plot(x,y, linestyle='dashed', color='blue')

plt.subplot(1, 2, 2)
plt.plot(x,z, color='red')
plt.show()
```

**See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
fig = plt.figure()
plt.subplots(nrows=1, ncols=2, figsize=(8, 2))
plt.subplot(1, 2, 1)
plt.plot(x,y, color='blue', linewidth='4')

plt.subplot(1, 2, 2)
plt.plot(x,z, linestyle='dashed', color='red', linewidth='4')
plt.tight_layout()
plt.show()
```
