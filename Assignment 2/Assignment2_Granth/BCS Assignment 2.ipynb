```python
# Importing the required libraries
import yfinance as yf
import plotly.graph_objects as go
```

```python
# Selecting and reading the stock's data
ticker = yf.Ticker('TORNTPHARM.NS')
h_data  = ticker.history(period = '6mo')
print(h_data)
```

```python
# Plotting candlestick chart
fig = go.Figure(data = [go.Candlestick(x = h_data.index, open = h_data['Open'], high = h_data['High'], low = h_data['Low'], close = h_data['Close'])])
fig.update_layout(title = 'TORRENT PHARMA STOCK PRICE')
fig.show()
```
