```python
import yfinance as yf
import plotly.graph_objects as go
df = yf.download('TATAMOTORS.NS' , period ='6mo' , interval = '1d')
df.head()
```

```python
from os import close
fig = go.Figure(data=[

   go.Candlestick( x = df.index,
                  open = df['Open'],
                  low = df['Low'],
                  high = df['High'],
                  close = df['Close']
                   )
   ]
                )








fig.update_layout(title = 'tata motors' , xaxis_title = 'Date' , yaxis_title = 'Price')
fig.show()
```
